**Regularization in Machine Learning**

One of the major aspects of training your ML model is **avoiding overfitting**. The model will have a low accuracy if it is overfitting. This happens because your
model is trying too hard to capture the noise in your training dataset.

By noise we mean the data points that dont really represent the true properties of your data, but random chance.

**Regularization** constrains or shrinks the coefficient estimates towards zero. In other words,** discourages learning a more complex** or flexible model, so as to avoid the risk of overfitting.

#### **Some maths...**

A multiple linear regression model describes how a variable $y$ depends linearly on a number of predictor variables $X = [1, x_1, x_2, ..., x_n]$

A multiple linear regression model can be written as:

$$ y = b_0 + b_1 x_1 + b_2 x_2 + ... + b_n x_n $$

Note: more complex models may include also higher power of the variable x, i.e.

$$ y = b_0 + b_1 x_1 + b_2 x_2 + ... + b_n x_n + $$
$$ + b_{n+1} x_1^2 + b_{n+2} x_1 x_2 + ...$$

They are linear models because the variable y can still be written as a linear combination of the parameters b's. Confusingly, are also sometime called non-linear, as the regression curve is not a line.

With a compact notation the linear regression model can be written in the form

$$ y = \beta X + \epsilon $$

where $X, \beta$ and $\epsilon$ are vectors.

In linear algebra terms, the least-square parameter estimates of $\beta$ are the vectors that minimize

$\sum_{i=1}^n \epsilon_i^2 = \epsilon'\epsilon = (y - \beta X)'(y-\beta X)$

We want to find the best $\beta$ in the sense that the sum of the squared residual is minimized. If the residual is zero, then 

$$\hat y = \hat \beta X.$$

In order to minimize the residual, we get

$$-2X'( y - \hat \beta X) = 0$$
$$X'  y - X' X \hat \beta = 0$$
$$X' X \hat \beta = X'  y$$

Finally, 

$$\hat \beta = (X' X)^{-1} X'  y$$

gives us the formulation of the problem of "learning" the beta parameters in a problem of multiple linear regression

##### **Step 2**: Define least squares algorithm

First, we initialize the class in a general form. $\beta$ is a vector of the coefficients of the model. It can be accessed later, to be visualized for example.

````python
class MultipleLinearRegression(object):
    
    def __init__(self):
        
        self.beta = [] # coefficients
````

Now, we define a method in order to reshape the feature matrix $X$. In particular, 
- $X$ must be a bi-dimensional matrix (number of samples $\times$ number of features) 
- a first column of ones must be concatenated to perform the linear algebra.

````python
class MultipleLinearRegression(object):
    
    def __init__(self):
        
        self.beta = [] # coefficients
        
    def _reshape_input(self, X):
        
        # If X is not bi-dimensional (case of a single feature), reshape it
        if len(X.shape) == 1:
            X = X.reshape(-1, 1)
        
        # Then, concatenate a column of ones
        ones = np.ones((X.shape[0], 1))
        X = np.hstack((ones, X))
        
        return X
        
````

Now, we are to define the "heart" of our model, namely:
- a method fit(), which will be responsible for training the model parameters
- a method predict(), which will be useful for the end-user in order to validate the model and make new predictions.

The __fit()__ method include the formula determined before, to calculate the $\beta$ coeffiecients for our training data. The fit() method will also take care of reshaping the input data. 

````python
class MultipleLinearRegression(object):
    
    def __init__(self):
        
        self.beta = [] # coefficients
        
    def _reshape_input(self, X):
        
        # If X is not bi-dimensional (case of a single feature), reshape it
        if len(X.shape) == 1:
            X = X.reshape(-1, 1)
        
        # Then, concatenate a column of ones
        ones = np.ones((X.shape[0], 1))
        X = np.hstack((ones, X))
        
        return X
        
    def fit(self, X, y):
        
        X = _reshape_input(X)
        
        #theta = y * X.T * inv(X*X.T)
        self.beta = np.linalg.inv(X.transpose().dot(X)).dot(X.transpose()).dot(y)
````

The method __predict()__, instead, consists of a forward computation which calculates the output of the model on a given batch of samples.

````python
class MultipleLinearRegression(object):
    
    def __init__(self):
        
        self.beta = [] # coefficients
        
    def _reshape_input(self, X):
        
        # If X is not bi-dimensional (case of a single feature), reshape it
        if len(X.shape) == 1:
            X = X.reshape(1, -1)
        
        # Then, concatenate a column of ones
        ones = np.ones((X.shape[0], 1))
        X = np.hstack((ones, X))
        
        return X
        
    def fit(self, X, y):
        X = self._reshape_input(X)
        self.beta = np.linalg.inv(X.transpose().dot(X)).dot(X.transpose()).dot(y)
        
    def predict(self, X):
        
        X = self._reshape_input(X)
        
        if len(self.beta) != X.shape[1]:
            print("Coefficients and input data size do not fit.", len(self.beta), X.shape)
            return None
        
        else:
            return X.dot(self.beta)
````

````python
def mean_squared_error(y, y_pred):
    
    return np.mean((y_pred - y)**2)
````

````python
mean_squared_error(y_test, predicitons)
````

#### **Step 3**: $R^2$ score

Compute a score to evaluate how well a linear model fit on our data.

The coefficient $R^2$, said "R squared", is the proportion of the variance in the dependent variable that is predictable from the independent variable(s).

It is defined as $R^2 = 1 - \frac{u}{v}$, 
- $u$ is the residual sum of squares ((y - y_pred) ** 2).sum() 
- $v$ is the total sum of squares ((y - y.mean()) ** 2).sum(). 

Another definition is: $R^2 = \frac{total-variance-explained-by-model}{total-variance}$

The best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse). 

````python
r2_train = lrm.score(X_train, y_train)
r2_test = lrm.score(X_test, y_test)

print("The r^2 score: {} (TRAIN) {} (TEST)".format(r2_train, r2_test))
````

### **Ex_20: The importance of Regularization**

Different regularization techniques can be used for linear model in order to prevent overfitting. Among them,
- Lasso
- Ridge
- Elastic-Nets

All of this model can be easily imported with `sklearn`.

#### **Lasso regression**

It addresses some of the problems of Ordinary Least Squares by imposing a penalty on the size of the coefficients. (Linear least squares with l1 regularization.)

The ridge coefficients minimize a penalized residual sum of squares:

$$\min_{\beta} \vert\beta X - y\vert ^ 2  + \alpha \vert\beta\vert $$

Informally, "I want to fit a line on my input features but ALSO find unuseful features".

The value of $\alpha$ controls the amount of shrinkage:

- the larger the $\alpha$, the more zero-parameters.

This raise a good effect: it is a way to perform feature selection.

Note: we indicate with $\beta$ a vector of coefficients while $\alpha$ is a single hyper-parameter.

#### **Ridge regression** 

It addresses some of the problems of Ordinary Least Squares by imposing a penalty on the size of the coefficients. (Linear least squares with l2 regularization.)

The ridge coefficients minimize a penalized residual sum of squares:

$$\min_{\beta} \vert\beta X - y\vert ^ 2  + \alpha \vert\beta\vert ^ 2 $$

Informally, "I want to fit a line on my input features but ALSO find small values for the $\beta$ parameters".

The value of $\alpha$ controls the amount of shrinkage:

- the larger the $\alpha$, the greater the amoun of shrinkage.

This raise a good effect: coefficients become more robust to collinearity (when independent variables in the regression model are highly correlated to each other).

#### **Elastic-Nets**

It addresses some of the problems of Ordinary Least Squares by imposing the combination of the two previously defined penalties on the size of the coefficients. (Linear least squares with l1 and l2 regularization.)

The ridge coefficients minimize a penalized residual sum of squares:

$$\min_{\beta} \vert\beta X - y\vert ^ 2  + \alpha_1 \vert\beta\vert + \alpha_2 \vert\beta\vert ^ 2 $$

Informally, "I want to fit a line on my input features but ALSO find unuseful features but ALSO find small values for the $\beta$ parameters".

##### **Step 3**: Grid search cross validation

How can we determine the best hyper-parameters ($\alpha$, ...) for our model?

An option is to try as more combinations as possible, evaluate each of them and then selecting the best estimator.

Sklearn provides a utility to implement grid search over the parameters space which does exactly that.

**Grid search**

To perform an exhaustive search over specified parameter values for an estimator we perform grid search. There exists a sklearn class to perform grid seach over a number of hyper-parameters.

`GridSearchCV()`

It need two mandatory arguments:

- `estimator`: an estimator object. This is assumed to implement the scikit-learn estimator interface. Either estimator needs to provide a score function, or scoring must be passed.

- `param_grid`: dict or list of dictionaries. Dictionary with parameters names (string) as keys and lists of parameter settings to try as values, or a list of such dictionaries, in which case the grids spanned by each dictionary in the list are explored. This enables searching over any sequence of parameter settings.


````python
# Define the estimator
elasticnet = ElasticNet()

# Have a look at the default hyper-parameters
elasticnet.get_params()
````

````python
# Define a parameter grid to test
parameters = {
    'alpha': np.linspace(.1, 100., 100), 
    'l1_ratio': np.linspace(.1, 1., 10)
    }
````

````python
# Define the dataset

# Noisy dataset
X_train, X_test, y_train, y_test = train_test_split(X_noisy, y, test_size=.2, shuffle=True)
````

````python
# Define 
from sklearn.model_selection import GridSearchCV
gs = GridSearchCV(elasticnet, parameters, cv=5)

gs.fit(X_train, y_train)
````

````python
# Access gird search results
sorted(gs.cv_results_.keys())
````

````python
# Maximum test score (AVERAGE AMONG FOLDS)
gs.cv_results_['mean_test_score'].max()
````

````python
# Get the best configuration of hyper-parameters
gs.best_params_
````

````python
# Set the best estimator
elasticnet = gs.best_estimator_
````
